<h1> Seaborn regression plots usages </h1>

````
import seaborn as sns
%matplotlib inline
````

Loading seaborn <strong>dataset</strong>

````
tips = sns.load_dataset('tips')
````

````
tips.head()
````

Creating linear regression plot using
```
sns.lmplot
```


````
sns.lmplot(x = "total_bill", y="tip", data = tips, hue="sex", col = "day", palette="magma",  )
````
